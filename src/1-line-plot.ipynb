### Exercise 1: Line Plot Customization

```python
#
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = x**2  # y values are the square of x

# Create the plot
plt.plot(x, y, label='y = x^2', color='b')  # Blue line

# Customize the plot
plt.title('Plot of y = x^2')  # Title of the plot
plt.xlabel('x values')  # x-axis label
plt.ylabel('y values')  # y-axis label
plt.grid(True)  # Add grid
plt.legend()  # Show legend

# Show the plot
plt.show()
```
